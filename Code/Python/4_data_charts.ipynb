```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy.stats as st
```

```python
songs_csv = "../../Resources/merged_songs_data.csv"
songs_df = pd.read_csv(songs_csv, index_col=0)
songs_df.head()
```

```python
# Danceability vs Energy
plt.scatter(songs_df.iloc[:,1],songs_df.iloc[:,2], s=1, alpha=0.1)
plt.xlabel('Danceability')
plt.ylabel('Energy')
plt.savefig('../../Images/energy_v_danceability.png',dpi=100)
plt.show()
```

```python
# Accousticness vs Liveness
plt.scatter(songs_df.iloc[:,6],songs_df.iloc[:,8], s=1, alpha=0.1)
plt.xlabel('Acousticness')
plt.ylabel('Liveness')
plt.savefig('../../Images/accousticness_v_liveness.png',dpi=100)
plt.show()
```

```python
# Danceability vs Valence
plt.scatter(songs_df.iloc[:,2],songs_df.iloc[:,7], s=1, alpha=0.1)
plt.xlabel('Accousticness')
plt.ylabel('Energy')
plt.savefig('../../Images/accousticness_v_energy.png',dpi=100)
plt.show()
```

```python
# Danceability vs Valence
plt.scatter(songs_df.iloc[:,4],songs_df.iloc[:,7], s=1, alpha=0.1)
plt.xlabel('Loudness')
plt.ylabel('Energy')
plt.savefig('../../Images/loudness_v_energy.png',dpi=100)
plt.show()
```

```python
import seaborn as sns
corr = songs_df.corr()
sns.heatmap(corr, 
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values)

plt.savefig('../..Images/correlations_2.png')
```
