```python
import pandas as pd
```

```python
songs_csv = "../../Resources/songs_with_popularity_trimmed.csv"
songs_df = pd.read_csv(songs_csv)
songs_df.head()
```

```python
# Create new column with binary category label - 0 popularity or not

def bin_popularity (row):
    if row['popularity'] == 0:
        return '0'
    else:
        return '1'
```

```python
songs_df['bin_popularity'] = songs_df.apply (lambda row: bin_popularity(row), axis=1)
```

```python
songs_df.head()
```

```python
songs_df.to_csv('../../Resources/songs_with_popularity_trimmed_bins.csv',index=False)
```
