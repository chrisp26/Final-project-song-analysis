```python
import pandas as pd
```

```python
# Import CSVs and create dataframes

songs_csv = "../../Resources/spotify_tracks_features.csv"
pop_csv = "popularity.csv"

songs_df = pd.read_csv(songs_csv)
songs_2 = songs_df[['id', 'danceability', 'energy',
       'key', 'loudness', 'mode', 'speechiness', 'acousticness',
       'instrumentalness', 'liveness', 'valence', 'tempo', 'duration_ms',
       'time_signature']]

pop_df = pd.read_csv(pop_csv)
pop_df = pop_df['popularity']

pop_df.head()
# songs_2.head()
```

```python
# Merge dataframes
merged_df = pd.concat([songs_2, pop_df], axis=1)
merged_df.head()
```

```python
# Export dataframes
merged_df.to_csv('../../Resources/merged_songs_data.csv')
```
